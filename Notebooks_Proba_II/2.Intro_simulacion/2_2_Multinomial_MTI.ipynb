## Simulaciones de una Multinomial

Haremos el mismo caso de arriba, pero para una función con 3 variables aleatorias

```python
# Simulación de la categórica
p1 = 0.2
p2 = 0.3
p3 = 1-p1-p2

N_exp= 10

X_1 = 0
X_2 = 0
X_3 = 0


for _ in range(N_exp):
    u=np.random.random()

    if u<p1:
        X_1 = X_1 +1
    elif u<p1+p2:
        X_2 = X_2 +1
    else:
        X_3 = X_2 +1

print(X_1)
print(X_2)
print(X_3)
```

Ahora ponemos la multinomial de 3 variables en una función y ejecutamos muchas veces

```python
def sim_multi_3(p1,p2,N_exp):
    X_1 = 0
    X_2 = 0
    X_3 = 0


    for _ in range(N_exp):
        u=np.random.random()

        if u<p1:
            X_1 = X_1 +1
        elif u<p1+p2:
            X_2 = X_2 +1
        else:
            X_3 = X_3 +1
    
    return np.array([X_1, X_2, X_3])

sim_multi_3(0.2,0.5,10)
```

Hacemos muchas repeticiones de este experimento

```python
resultados = []

for i in range(N_sim):
    resultados.append(sim_multi_3(0.2, 0.5, 10))

resultados
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

resultados_array= np.array(resultados)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') 

# Plot the scatter plot
ax.scatter(resultados_array[:,0],resultados_array[:,1] ,resultados_array[:,2] , c='b', marker='o')

# Add labels and show the plot
ax.set_xlabel('$X_1$')
ax.set_ylabel('$X_2$') 
ax.set_zlabel('$X_3$')
plt.show()
```
